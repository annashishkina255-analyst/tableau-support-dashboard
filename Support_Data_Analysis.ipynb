```python
# Импортируем библиотеки
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Настраиваем случайность (для воспроизводимости)
np.random.seed(42)

# Создаем период: последние 90 дней
dates = [datetime.now() - timedelta(days=x) for x in range(90)]
dates.reverse()

# Создаем искусственные данные
data = {
    'ticket_id': range(1000, 1000 + len(dates)),
    'created_date': dates,
    'category': np.random.choice(['Network', 'Software', 'Hardware', 'Login', 'Database'], len(dates)),
    'priority': np.random.choice(['Low', 'Medium', 'High', 'Critical'], len(dates), p=[0.4, 0.3, 0.2, 0.1]),
    'agent': np.random.choice(['Agent_A', 'Agent_B', 'Agent_C', 'Agent_D'], len(dates)),
    'sla_target': np.random.choice([4, 8, 24, 72], len(dates)), # часы на решение
    'time_to_resolve': np.abs(np.random.normal(10, 8, len(dates))).round(1), # случайное время решения
    'status': np.random.choice(['Resolved', 'Closed'], len(dates))
}

# Создаем DataFrame
df = pd.DataFrame(data)

# На основе времени решения и SLA目标 определяем, нарушен ли SLA
df['sla_breached'] = df['time_to_resolve'] > df['sla_target']

# Сохраняем в CSV файл
df.to_csv('support_tickets_data.csv', index=False, encoding='utf-8-sig')

# Показываем первые 5 строк
print("Файл 'support_tickets_data.csv' создан! Вот как он выглядит:")
df.head()
```
